```python
from defect_gen import generate_dataset, show_samples

images, labels = generate_dataset(n_samples=200, size=8, defect_rate=0.10,
                              normal_range=(0.3, 0.5), defect_range=(0.95, 0.99),
                              seed=7)
# show_samples(images, labels, n_show=30) 
print(images[0], labels[0])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    images, labels,
    test_size=0.20,
    random_state=0,
    stratify=labels,     # <-- 80% of healthy AND 80% of defective go to train
)

print("train:", X_train.shape[0], "healthy:", (y_train==0).sum(), "defect:", (y_train==1).sum())
print("test :", X_test.shape[0],  "healthy:", (y_test==0).sum(),  "defect:", (y_test==1).sum())
```
